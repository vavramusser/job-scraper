```python
import requests
from bs4 import BeautifulSoup

url = "https://www.phi.org/employment/current-opportunities"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
}

response = requests.get(url, headers=headers, timeout=10)
soup = BeautifulSoup(response.content, 'html.parser')

# Save the HTML to a file so we can inspect it
with open('phi_page.html', 'w', encoding='utf-8') as f:
    f.write(soup.prettify())

print("HTML saved to phi_page.html")
print("\nLooking for common job listing patterns...")

# Try to find links that might be jobs
all_links = soup.find_all('a')
print(f"\nFound {len(all_links)} total links on page")

# Look for anything that might be a job listing
potential_jobs = []
for link in all_links:
    href = link.get('href', '')
    text = link.get_text(strip=True)
    if text and len(text) > 10:  # Jobs usually have meaningful text
        print(f"\nLink text: {text[:100]}")
        print(f"URL: {href}")
        potential_jobs.append((text, href))

print(f"\n\nFound {len(potential_jobs)} potential job listings")
```
